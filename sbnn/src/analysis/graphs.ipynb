```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Nested', 'Overlapping', 'Flat', 'Overall']
tees = [42.70, 34.49, 56.81, 52.16]
sbnn = [45.24, 36.92, 60.5, 54.36]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
fig.set_size_inches(9, 6)
rects1 = ax.bar(x - width/2, tees, width, label='TEES')
rects2 = ax.bar(x + width/2, sbnn, width, label='SBNN k=8')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('F1-Score (%)')
ax.set_title('F1 Scores by category')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_xlabel("Category")
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
